# Scatter Plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color="r")
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","m","r","y"]
sizes=[50,200,300,400,500,600,700]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.7,marker="*",edgecolor="g",linewidth=2)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[2,9,7,5,9,8,1]
colors=[10,49,30,29,56,20,39]
sizes=[50,200,300,400,500,600,700]
plt.scatter(day,no,c=colors,s=sizes,cmap="BrBG")
plt.scatter(day,no2,c="r",s=sizes)
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```
