# Bar Plot

```python
import matplotlib.pyplot as plt
x=["python","C","C++","java"]
y=[85,70,60,82]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Best Programming Languages",fontsize=15)
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="r",linewidth=5,linestyle=":")
plt.show()
```
