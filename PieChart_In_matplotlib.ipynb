# Pie Chart 

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["c","c++","Java","Python"]


plt.pie(x,labels=y)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["c","c++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["c","c++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["c","c++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.2,startangle=0
       ,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":5,"edgecolor":"m"}
       ,center=[3,3])

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["c","c++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.9,labeldistance=1.2,startangle=0
       ,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":5},rotatelabels=False)

plt.title("Programming Languages")
plt.legend(loc=2)
plt.show()
```

```python
# Donate Pie Chart
import matplotlib.pyplot as plt

x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","Java","Python"]

c=["r","b","g","y"]

plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=0.5,colors=c)


plt.show()
```
