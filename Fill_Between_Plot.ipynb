# Fill Between Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4),alpha=0.4)

plt.plot(x,y,color="r")
plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.show()
```
