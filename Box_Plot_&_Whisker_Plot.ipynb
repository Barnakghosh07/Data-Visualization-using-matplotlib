# Box Plot

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]


plt.boxplot(x,notch=True)


plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]


plt.boxplot(x,vert=True,widths=0.2,labels=["Python"],patch_artist=True,
            showmeans=True)


plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]


plt.boxplot(x,vert=True,widths=0.2,labels=["Python"],
            showmeans=True,sym="g+")


plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]


plt.boxplot(x,vert=True,widths=0.2,labels=["Python"],
            showmeans=True,sym="g+",
            boxprops=dict(color="r"),capprops=dict(color="b"),
           flierprops=dict(markeredgecolor="b"))


plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]
x1=[10,20,30,40,70,90,100]

y=[x,x1]
plt.boxplot(y,labels=["Python","C++"],
            showmeans=True,sym="g+",
            boxprops=dict(color="r"),capprops=dict(color="b"),
           flierprops=dict(markeredgecolor="b"))


plt.show()
```
